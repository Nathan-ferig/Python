```python
# Import libs

# Libs for modeling
import numpy as np
import pandas as pd

# Libs for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Lib to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read data
data = pd.read_excel('Dados_Pib.xlsx')

data.shape
```

```python
data.head()
```

```python
data.groupby( by = ['Territorialidades','Ano']).mean()
```

```python
data.rename(
    columns ={
        'Territorialidades' : 'Território'
    },
    inplace = True
);
```

```python
# Report grid

# Background color
color = 'lightgrey'

# Grid
grid_graph = sns.FacetGrid( data, col='Território', hue='Território', col_wrap=4)

# Add a graph in each available grid
grid_graph = grid_graph.map( plt.plot, 'Ano', 'PIB per capita')

# Edit design
grid_graph = grid_graph.map( plt.fill_between, 'Ano', 'PIB per capita',alpha=0.1).set_titles('{col_name}')

# Subtitle
grid_graph = grid_graph.fig.suptitle('''
Brazil's PIB per Capita per State", 
This report was made as a training pratice for "Python para Análise de Dados"
The forementioned online course can be found at Udemy.
''', fontsize = 18)
plt.subplots_adjust( top = 0.9)
```
