```python
# import libs

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
# read data
data = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')

#dimension
data.shape
```

```python
# Head

data.head()
```

```python
# Info

data.info()
```

```python
# Describe

data.describe()
```

```python
# Time series

dataset = data.set_index('Data')

dataset.head()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))
plt.plot( dataset.index,dataset['Fechamento'])
plt.title('Stock Analysis - Magalu', fontsize=14, loc='left')

plt.xlabel('Time Series')
plt.ylabel('Stock Value (BRL)')
```

```python
moving_average = dataset['Fechamento'].rolling(5).mean()
moving_tendency = dataset['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))

plt.plot( dataset.index,dataset['Fechamento'])
plt.plot( dataset.index,moving_average)
plt.plot( dataset.index,moving_tendency)

plt.xlabel('Time Series')
plt.ylabel('Stock Value (BRL)')
```

```python
sns.boxplot(data=dataset,x='Fechamento')
```

```python
# monthly boxplot

data['Mes'] = data['Data'].dt.month

plt.figure( figsize=(16,5))
sns.boxplot(data=data, x='Mes', y='Fechamento')
```

```python
data.groupby(['Mes']).describe()['Fechamento']
```

```python
graph = go.Figure(
    data = [
            go.Candlestick(
                x = dataset.index,
                open = dataset['Abertura'],
                high = dataset['Maior'],
                low = dataset['Menor'],
                close = dataset['Fechamento']
            ) 
    ]
)

graph.show()
```
