```python
# Import libs

# Libs for modeling
import numpy as np
import pandas as pd

# Libs for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Lib to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read data
data = pd.read_excel('Dados_Empresas_Energia.xlsx')

data.shape
```

```python
data.head()
```

```python
# Graph style
plt.style.use('seaborn-darkgrid')

# Size
plt.figure( figsize=(16, 7))

# Title
plt.title('Energy sector evaluation analysis', loc='left', fontsize=18, fontweight=0 )

# Petrobras
plt.plot( data.index, data['Petrobras'], color='green', linewidth=4, alpha=0.7 )
# Text
plt.text( data.index[-1], data['Petrobras'].tail(1), 'Petrobras', color='green', size='large', horizontalalignment='left' )

# Other companies
for Coluna in data.columns[1:]:
  # Companies
  plt.plot( data.index, data[Coluna], color='gray', linewidth=2, alpha=0.7 )
  # Text
  plt.text( data.index[-1], data[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left' )

# Labels
plt.xlabel('Period')
plt.ylabel('Closing Price (R$)');
```
