```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
px.data.gapminder()
```

```python
df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
plt.plot(df.index, df['gdpPercap'])
plt.title("Brazil's GDP per Capita")
plt.ylabel("GDP per capita")
plt.xlabel("Year")
```

```python
title = "Life expectancy versus GDP per Capita in Brazil"

plt.figure(figsize = (12,4))
plt.plot(df['lifeExp'], df['gdpPercap'], '--bo')
plt.title(title, loc = 'left')
plt.ylabel("GDP per capita'")
plt.xlabel("Life Expectancy")
plt.show
```

```python
plt.bar(x = df.index,
       height = df['pop'],
       color = 'red'
       )
plt.title("Brazil's Population Growth")
```

```python
def filter_continent(continent):
    df = px.data.gapminder()
    df = df[df['continent'] == continent]
    return df
```

```python
def filter_country(country,variable):
    df = px.data.gapminder()
    df = df[df['country'] == country][variable]
    return df
```

```python
continent_name = 'Americas'
continent = filter_continent(continent_name)
countries = continent['country'].unique()
plt.figure(figsize = (12,8))
for country in countries:
    plt.scatter(filter_country(country = country,
                             variable = 'lifeExp'
                            ),
                filter_country(country = country,
                            variable = 'gdpPercap'
                            )
            )
title = 'Life expectancy versus GDP per Capita in ' + continent_name
plt.title(title, loc = 'left')
plt.legend(labels = countries, loc = 'best')
plt.ylabel("GDP per capita'")
plt.xlabel("Life Expectancy")
```
